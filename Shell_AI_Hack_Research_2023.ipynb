## From the above, we see that the Kharif season, which is usually characterized with heavy rains and high humidity was the highest cultivation period for cotton. The Kharif season in India refers to the summer monsoon season, which typically starts in June and lasts until September. It is a crucial agricultural period for the country, as the majority of India's rainfall occurs during this time, providing essential water for crops. The Kharif season is characterized by warm temperatures, high humidity, and heavy rainfall.

## The plot above shows that the highest production of cotton is in the Kharif season. So was there an alteration of this season in 2012 and 2015/2016??? Based on external research, there was a fluctuation of rains in this season for 2015/2016 and 2018/2019, which caused a reduction in the production quantity, and subsequently, a reduction in the quantity of biomass gotten from cotton.
